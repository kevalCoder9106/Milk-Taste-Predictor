```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sb

sb.set()
```

```python
raw_data = pd.read_csv('milknew.csv')
raw_data.describe(include='all')
```

```python
# removing Grade observation, since its not necessary
data = raw_data.copy()
data = data.drop(['Grade'], axis=1)

data.head()
```

```python
def config_and_split_data(data, test_size):
    # define inputs and targets
    inputs = data.drop(['Taste'], axis=1)
    target = data['Taste']

    # adding const
    const_inputs = sm.add_constant(inputs)

    # splitting the data
    from sklearn.model_selection import train_test_split
    return (train_test_split(inputs,target,test_size=test_size,random_state=42))
```

```python
x_train, x_test, y_train, y_test = config_and_split_data(data,.05)
```

```python
# training model
reg_log = sm.Logit(y_train,x_train)
model = reg_log.fit()

pred_res = model.pred_table()
accuracy = (((pred_res[0,0] + pred_res[1,1]) / pred_res.sum()) * 100).round(3)

print('Accuracy of the model is ' + str(accuracy) + ' %')
```

```python
# Model summary
model.summary()

# From below summary we can identify that pH, Turbidity, Colour are insignificant observation. We will re-train the model after testing this model.
```

## Predicted values 

### 0: Bad taste, 1: Good taste

```python
# testing the model
pred_values = model.predict(x_test).round(0)

prediction_table = pd.DataFrame({'Original values': y_test, 'Predicted values': pred_values})
prediction_table
```

### Prediction analysis using confusion matrix

```python
confu_matrix = model.pred_table()
confu_matrix

matrix_table = pd.DataFrame(confu_matrix)
matrix_table.columns = ['Predicted 0', 'Predicted 1']
matrix_table = matrix_table.rename(index={0: 'Actual 0', 1: 'Actual 1'})
matrix_table

# As we can see our model is more confused in predicting 0 incorrect
```

# Accuracy of the model is 66.499 %

## Phase 2
#### Since  pH, Turbidity, Colour are insignificant observation according to our old model, we will analyse pH and drop Turbidity and Colour.

```python
new_data = data.drop(['Turbidity','Colour'],axis=1)
```

### Removing outliers
#### pH

```python
sb.displot(new_data['pH'])
```

#### As we can see there are few outliers, so in this phase we will remove few outliers and drop Turbidity and Colour observations

```python
q_ph = new_data['pH'].quantile(.03)
new_data = new_data[new_data['pH'] > q_ph]

q_ph = new_data['pH'].quantile(.98)
new_data = new_data[new_data['pH'] < q_ph]
```

```python
sb.displot(new_data['pH'])
```

### Removing outliers
#### Temprature

```python
sb.displot(new_data['Temprature'])
```

```python
q_tempr = new_data['Temprature'].quantile(0.99)
new_data = new_data[new_data['Temprature'] < q_tempr]
```

```python
sb.displot(new_data['Temprature'])
```

```python
# splitting data

x_train, x_test, y_train, y_test = config_and_split_data(new_data,0.05)
```

```python
# training model

reg_log2 = sm.Logit(y_train,x_train)
model2 = reg_log2.fit()

pred_tab = model2.pred_table()
accuracy2 = ((pred_tab[0,0] + pred_tab[1,1]) / pred_tab.sum()) * 100

accuracy2
```

```python
# model summary

model2.summary()
```
